# working with dates

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

 EXTRACT YEAR

```python
date['date_year'] =  date['date'].dt.year
date.sample(5)
```

EXRTACT MONTH

```python
date['date_month_no'] =  date['date'].dt.month
date['date_month_name'] =  date['date'].dt.month_name()

date.sample(3)
```

EXTRACT DAYS

```python
date['date_day'] =  date['date'].dt.day

# day of week 
date['date_day_ofweek'] =  date['date'].dt.dayofweek

# day of week name 

date['date_dayofweek_name'] =  date['date'].dt.day_name()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dayofweek_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()

```

Extract week of the year

```python
date['date_week'] = date['date'].dt.weekday
date.drop(columns=['product_id','city_id','orders']).sample(15)
```

EXTRACT QUARTER

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

EXTRACT TIME ELAPSED BETWEEN DATES

```python
import datetime
today = datetime.datetime.today()

print(today)
```

```python
# calc how many days have passed from today  

(today - date['date']).dt.days
```

```python
# Months passed

np.round((today -date['date']) / np.timedelta64(1, 'W'),0)/4
```
